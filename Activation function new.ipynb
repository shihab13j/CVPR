```python
import matplotlib.pyplot as plt
import numpy as np

def elu(x, alpha=1.0):
    return np.where(x >= 0, x, alpha * (np.exp(x) - 1))

x = np.arange(-10, 11, 1)

```

```python
y = elu(x)
plt.plot(x,y, 'o--')
plt.grid(True)
plt.xlabel('X')
plt.ylabel('ELU(X)')
plt.title('ELU Activation Function')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Step function
def step(x):
    return np.where(x >= 0, 1, 0)

# Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# TanH function
def tanh(x):
    return np.tanh(x)

# ReLU function
def relu(x):
    return np.maximum(0, x)

# PReLU function
def prelu(x, alpha=0.1):
    return np.where(x >= 0, x, alpha * x)

# Define the input range
x = np.arange(-10, 11, 1)

# Plot the activation functions
plt.plot(x, step(x), label='Step')
plt.plot(x, sigmoid(x), label='Sigmoid')
plt.plot(x, tanh(x), label='TanH')
plt.plot(x, relu(x), label='ReLU')
plt.plot(x, prelu(x), label='PReLU')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Activation Functions')
plt.show()
```

# Step function:
The Step function has a binary output: it returns 1 for input values greater than or equal to 0, and 0 otherwise. It is a simple and efficient way to introduce nonlinearity into a neural network, and is commonly used in binary classification problems. However, the Step function suffers from the "vanishing gradient" problem, where gradients become very small or zero for input values far from the decision boundary.

# Sigmoid function:
The Sigmoid function has an output between 0 and 1, and is commonly used as a "squashing" function to convert any input into a probability value. It has a smooth gradient that makes it well-suited for backpropagation in a neural network, but it also suffers from the vanishing gradient problem for input values far from the origin.



# TanH function:
The TanH function is similar to the Sigmoid function, but it has an output between -1 and 1. Like the Sigmoid function, it is smooth and differentiable, but it suffers from the vanishing gradient problem for large input values.



# ReLU function:
The ReLU (Rectified Linear Unit) function returns 0 for negative input values, and the input value itself for positive input values. It is a simple and efficient way to introduce nonlinearity into a neural network, and has been shown to work well in practice for many tasks. It does not suffer from the vanishing gradient problem, and is computationally efficient to compute. However, ReLU can also suffer from the "dying ReLU" problem, where the gradient becomes zero for negative input values and the neuron effectively "dies".



# PReLU function:
The PReLU (Parametric Rectified Linear Unit) function is similar to ReLU, but it introduces a learnable parameter that determines the slope of the function for negative input values. This can help to address the dying ReLU problem, and has been shown to work well in practice for some tasks.



# ELU function:
The ELU (Exponential Linear Unit) function is similar to the ReLU function for positive input values, but it introduces a negative slope for negative input values. This helps to address the dying ReLU problem and can improve the performance of the network. ELU also has a smooth gradient that makes it well-suited for backpropagation, and does not suffer from the vanishing gradient problem.

In summary, each activation function has its advantages and disadvantages, and the choice of which to use depends on the specific task and architecture of the neural network. ReLU is a good default choice due to its simplicity and efficiency, but PReLU and ELU can be better choices in some cases. Sigmoid and TanH are less commonly used now due to their vanishing gradient problems.

# Comparison of all Activation Function's uses
The choice of activation function depends on the specific task and the characteristics of the data being modeled. Here are some general guidelines on which activation function works better than others for different sections:

Binary Classification:
For binary classification problems, the sigmoid activation function is commonly used. It outputs values between 0 and 1, which can be interpreted as probabilities. The step function can also be used for binary classification, but it is not differentiable, which can cause problems during training.

Multi-class Classification:
For multi-class classification problems, the softmax activation function is commonly used. It outputs a probability distribution over the different classes, which can be useful for making predictions. Other activation functions that can be used for multi-class classification include the sigmoid and tanh functions.

Regression:
For regression problems, the linear activation function is commonly used. It outputs the input value multiplied by a weight, which can be useful for predicting continuous values. Other activation functions that can be used for regression include the sigmoid, tanh, and ReLU functions.

Hidden Layers:
For hidden layers, the ReLU activation function is commonly used. It is computationally efficient and has been shown to work well in many applications. However, it can suffer from the "dying ReLU" problem, where some neurons "die" and stop outputting anything other than 0. In these cases, the Leaky ReLU or PReLU activation function can be used to address this issue.

Vanishing Gradient Problem:
The vanishing gradient problem occurs when the gradients become too small to update the weights in the earlier layers of the network. This can occur when using activation functions that saturate, such as the sigmoid and tanh functions. In these cases, the ReLU or Leaky ReLU activation function can be used instead.

Exploding Gradient Problem:
The exploding gradient problem occurs when the gradients become too large to update the weights in the network. This can occur when using activation functions that have unbounded output, such as the ReLU function. In these cases, the tanh or sigmoid activation function can be used instead.

In summary, there is no one-size-fits-all activation function that is best for all situations. The choice of activation function depends on the specific task and the characteristics of the data being modeled. It is important to experiment with different activation functions and choose the one that works best for a particular problem.





```python
import matplotlib.pyplot as plt
import numpy as np

def selu(x, alpha=1.67326, scale=1.0507):
    return scale * np.where(x > 0, x, alpha * (np.exp(x) - 1))

x = np.arange(-10, 11, 1)
y = selu(x)

plt.plot(x, y, 'o--')
plt.xlabel('X')
plt.ylabel('SELU(X)')
plt.title('SELU Activation Function')
plt.grid(True)
plt.show()
```

# SELU Activation Function:
The Scaled Exponential Linear Unit (SELU) is a type of activation function used in neural networks that has been shown to produce significant improvements in training speed and accuracy in certain types of networks. It is particularly effective in deep neural networks, where it has been shown to outperform other activation functions such as ReLU, LeakyReLU, and ELU.

Advantages:

The SELU activation function can help alleviate the vanishing gradient problem in deep neural networks, which is a common issue that can occur during training when the gradient becomes too small to be effective.
It has been shown to improve the performance of deep neural networks on a variety of tasks, including image classification, natural language processing, and speech recognition.
The activation function is self-normalizing, which means that the output of the function has a mean and variance that remain constant throughout the layers of the network. This can help speed up training and reduce the likelihood of overfitting.

Disadvantages:

One potential disadvantage of the SELU activation function is that it may not be suitable for all types of neural networks. In particular, it is designed for feedforward networks with dense connections, and may not be as effective in other types of networks such as convolutional neural networks or recurrent neural networks.
Another potential issue with the SELU activation function is that it requires careful initialization of the network weights and biases, which can be more complex than other activation functions.
